### Arrays

Arrays are commonly used in programming languages due to their computational efficiency. Although python doesn't natively have Arrays built in, it's introduced by the library Numpy. Lists can be used as arrays but there are some key differences you need to be aware of:

- Arrays are fixed in size - lists are dynamic.
- Arrays can only store data of the same type - lists can store different data types.
- Lists can be used as arrays - but not the other way round.
- Array are more memory-efficient.

````python
import numpy as np

data = [2,4,6,8,10]

myFirstArray = np.array(data)
print(myFirstArray) # Notice the similarity of the output to a list.
````

# RECAP/Cheat Sheet

````python
# Selection - if 

name = "Nirosh"

if name[0] == 'N':
    print("Your name starts with N")
elif name[0] == 'A':
    print("Your name starts with A")
else: 
    print("Not Sure what your name starts with")
````

````python
# While Loops - To continue doing things UNTIL A CRITERIA IS NO LONGER MET.
# "For As Long As"

age = 11

while age <18:
    print(f"You can't enter the bar because you are {age} years old. This is under 18")
    #while loops always need an increment! 
    age = age + 1 
````

````python
# FOR loops - To iterate THROUGH a FINITE list of items.

text = "This is a story about an elephant."

#Instantiating Variables
vowels = "AEIOUaeiou"
count = 0 

for letter in text: 
    if letter in vowels:
        count = count + 1

print(f"The number of vowels is {count}")
````

````python
# Functions

def anyFunctionName():
    print("Hello world")

anyFunctionName()
````

````python
# FOR AND WHILE LOOP ON A LIST

myList = ["Strawberry", "Mango", "Bannana", "Cherry", "Pineapple"]

for fruit in myList:
    if fruit == "Strawberry" or fruit == "Cherry":
        print(f"{fruit} is in season!")
    else:
        print(f"{fruit} is not in season")
````

````python
myList = ["Strawberry", "Mango", "Bannana", "Cherry", "Pineapple"]

for fruit in myList:
    if fruit in ["Strawberry", "Mango"]:
        print(f"{fruit} is in season!")
    else:
        print(f"{fruit} is not in season")
````

````python
recallItems = ["Strawberry", "Mango", "Bannana", "Cherry", "Pineapple"]

counter = 0

while counter < len(recallItems):
    print(f"This item: {recallItems[counter]} is recalled")
    counter += 1 
````

## 7. A brief look at classes (NEW)

"Classes provide a means of bundling data and functionality together. Creating a new class creates a new type of object, allowing new instances of that type to be made. Each class instance can have attributes attached to it for maintaining its state. Class instances can also have methods (defined by its class) for modifying its state."
- Source: https://docs.python.org/3/tutorial/classes.html

In simple terms, think of classes a  cookie cutter that allows you to create multiple versions of an object in a templated/standardised way.

````python
class Student():
    def __init__(self,name):
        self.name = name
        
    def sayHello(self):
        print(f"Hello {self.name}, welcome to your python class.")
````

````python
student1 = Student(name="Nirosh")
student1.sayHello()
````

````python
# Task: Modify the Student class to ask the user for their age and check that they are above 18. 
class Student_v2():
    def __init__(self,name):
        self.name = name
        self.age = int(input("Give me your age"))
        
    def sayHello(self):
        print(f"Hello {self.name}, welcome to your python class.")

    def checkAge(self):
        if self.age > 18: 
            print("Welcome, you're above 18")
        else:
            print("Go away")

student2 = Student_v2(name="Charles")
student2.checkAge()

                  


# Stretch: Look into class inheritence and what it means. 
##Try doing the task above by inheriting the Student() into a new class called Advanced_Student().
````
